```python
def highestfreq(s):
    s = s.split(" ")
    d = {}
    for i in s:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    max_ans = 0
    max_val = 0
    for i in s:
        if max_ans<d[i]:
            max_ans = max(max_ans,d[i])
            max_val = len(i)
    return max_val

print(highestfreq("write write write all the number from from from 1 to 100"))
print(highestfreq("write write all the number from from from 1 to 100"))
print(highestfreq("write write all all all the number from from  1 to 100"))
```

### Q2

```python
from collections import Counter
def valid(s):
    c = Counter(s)
    if len(s)==len(c):
        return "yes"
    ideal_freq = min(c.values())
#     print(c,len(c),ideal_freq)
    if(len(c)*ideal_freq == len(s) or len(c)*ideal_freq == len(s)-1):
        return "yes"
    else:
        return "no"
    
    
print(valid("abcc"))
print(valid("aaabbbcc"))
print(valid("harrsssha"))
print(valid("abc"))
```
